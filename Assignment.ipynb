# Create Visualizations

```python
import matplotlib.pyplot as plt
import seaborn as sns
popular_dishes.head(10).plot(kind='bar')
plt.title('Top 10 Popular Dishes')
plt.show()
```

# Plot the bar chart

```python

plt.figure(figsize=(8, 5))
sns.barplot(x=popular_dishes.values, y=popular_dishes.index, palette='viridis')
plt.title('Top 5 Dishes Ordered', fontsize=16)
plt.xlabel('Number of Orders', fontsize=12)
plt.ylabel('Dish Name', fontsize=12)
plt.show()
```

```python
# Plot the pie chart
plt.figure(figsize=(6, 6))
popular_dishes.plot.pie(autopct='%1.1f%%', colors=sns.color_palette('pastel'), startangle=90)
plt.title('popular_dishes', fontsize=16)
plt.ylabel('')  # Hides y-axis label
plt.show()
```
